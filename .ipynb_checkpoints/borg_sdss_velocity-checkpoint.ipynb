# Import data

```python
import numpy as np
velocity = np.load('borg_sdss_velocity.npz')

#Minimum and maximum position along the x-axis in Mpc/h
xmin=velocity['ranges'][0]
xmax=velocity['ranges'][1]

#Minimum and maximum position along the y-axis in Mpc/h
ymin=velocity['ranges'][2]
ymax=velocity['ranges'][3]

#Minimum and maximum position along the z-axis in Mpc/h
zmin=velocity['ranges'][4]
zmax=velocity['ranges'][5]

#3D probabilistic maps for velocity field
vx_mean=velocity['vx_mean']
vx_var=velocity['vx_var']
vy_mean=velocity['vy_mean']
vy_var=velocity['vy_var']
vz_mean=velocity['vz_mean']
vz_var=velocity['vz_var']
```

# Example plot: mean

```python
from matplotlib import pyplot as plt
%matplotlib inline
f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True, figsize=(15,5))
ax1.imshow(vx_mean[:,:,128], origin='lower', extent=[ymin,ymax,zmin,zmax], cmap="magma")
ax1.set_title('$v_x$')
ax1.set_aspect('equal')
ax2.imshow(vy_mean[:,:,128], origin='lower', extent=[ymin,ymax,zmin,zmax], cmap="magma")
ax2.set_title('$v_y$')
ax3.imshow(vz_mean[:,:,128], origin='lower', extent=[ymin,ymax,zmin,zmax], cmap="magma")
ax3.set_title('$v_z$')
plt.show()
```

# Example plot: variance

```python
from matplotlib import pyplot as plt
%matplotlib inline
f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True, figsize=(15,5))
ax1.imshow(vx_var[:,:,128], origin='lower', extent=[ymin,ymax,zmin,zmax], cmap="viridis")
ax1.set_title('$v_x$')
ax1.set_aspect('equal')
ax2.imshow(vy_var[:,:,128], origin='lower', extent=[ymin,ymax,zmin,zmax], cmap="viridis")
ax2.set_title('$v_y$')
ax3.imshow(vz_var[:,:,128], origin='lower', extent=[ymin,ymax,zmin,zmax], cmap="viridis")
ax3.set_title('$v_z$')
plt.show()
```

```python
# Astropy imports
from astropy.coordinates import SkyCoord  # High-level coordinates
from astropy.coordinates import ICRS, Galactic, FK4, FK5  # Low-level frames
from astropy.table import Table,Column
```

```python
#Create 1D Table from 3D arrays, calc position vector from indices 
pos = np.zeros((256,256,256,3))
for i in range(256):
    pos[i,:,:,0] = (i/255.)* (xmax-xmin) +  xmin
    pos[:,i,:,1] = (i/255.)* (ymax-ymin) +  ymin
    pos[:,:,i,2] = (i/255.)* (zmax-zmin) +  zmin
dist = np.linalg.norm(pos,axis=3)
vmag = np.sqrt(vx_mean**2 + vy_mean**2 + vz_mean ** 2)
mask = dist <= 350.

velTable = Table(names=('x', 'y', 'z','vx','vy','vz','dist','vmag'), data = 
           [pos[mask,0].flatten(),pos[mask,1].flatten(),pos[mask,2].flatten(),
            vx_mean[mask].flatten(),vy_mean[mask].flatten(),vz_mean[mask].flatten(),
           dist[mask].flatten(),vmag[mask].flatten()])
```

```python
# convert position and velocity to galactic coordinates and add to table
posCoords=SkyCoord(x=velTable['x'],y=velTable['y'],z=velTable['z'],representation='cartesian')
velTable.add_column(Column(posCoords.galactic.cartesian.x,name="pos_x"),0)
velTable.add_column(Column(posCoords.galactic.cartesian.y,name="pos_y"),1)
velTable.add_column(Column(posCoords.galactic.cartesian.z,name="pos_z"),2)

velCoords=SkyCoord(x=velTable['vx'],y=velTable['vy'],z=velTable['vz'],representation='cartesian')
velTable.add_column(Column(velCoords.galactic.cartesian.x,name="vel_x"),3)
velTable.add_column(Column(velCoords.galactic.cartesian.y,name="vel_y"),4)
velTable.add_column(Column(velCoords.galactic.cartesian.z,name="vel_z"),5)
velTable
```

```python
# only keep nine columns for Uniview USES format, output ascii table
velTable.remove_columns(['y','z','vx','vy','vz'])
```

```python
#write table
fmtDict = {}
for c in velTable.colnames:
    fmtDict[c] = '%.4f'
velTable.write("sdss_borg_velocity.raw",format="ascii.no_header", formats = fmtDict)
```
